```python
import pandas as pd
import numpy as np
from os import listdir

import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as mpatches
from pylab import cm
import scipy.optimize


#from my_functions import find_min_distance
```

```python
def process_signal(df_int, df_bckgrd):
    df_int_ = df_int.copy()
    df_bckgrd_ = df_bckgrd.copy()
    return df_int_.Mean - df_bckgrd_.Mean
```

```python
directories = ["Baz_Jupiter_polar//", "Baz_Jupiter_nonpolar//", "Par1_Jupiter//", "Par1_Baz//", "Par1_Jupiter_long//", "aPKC_Baz_long//", "Par1_Baz_long//"]
timeintervals = [30, 30, 30, 60, 600, 600, 600]
green_channel = ["Baz", "Baz", "Par1", "Par1", "Par1", "aPKC", "Par1"]
red_channel = ["Jupiter", "Jupiter", "Jupiter", "Baz", "Jupiter", "Baz", "Baz"]

info_df = pd.DataFrame(data={"directory": directories, "timeinterval": timeintervals, "green_channel": green_channel, "red_channel": red_channel})
```

```python
directory = "Par1_Baz_long//"
```

```python
files = listdir(directory)
dt = info_df[info_df.directory==directory].timeinterval.values[0]

for index, file in enumerate(files[0::4]):
    first_channel = pd.read_csv(directory+file)
    second_channel = pd.read_csv(directory+files[index*4 + 2])
    first_channel_bckgrd = pd.read_csv(directory+files[index*4 + 1])
    second_channel_bckgrd = pd.read_csv(directory+files[index*4 + 3])
    # Create figure object and store it in a variable called 'fig'
    fig = plt.figure()

    # Add axes object to the figure that takes up entire figure
    ax = fig.add_axes([0, 0, 1, 1])
    process_signal(first_channel, first_channel_bckgrd).plot()
    process_signal(second_channel, second_channel_bckgrd).plot()
```
