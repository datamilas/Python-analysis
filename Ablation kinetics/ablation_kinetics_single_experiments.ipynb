```python
def process_signal(df_int, df_bckgrd, normalize_to_0=False):
    df_int_ = df_int.copy()
    df_bckgrd_ = df_bckgrd.copy()
    intensity = df_int_.Mean - df_bckgrd_.Mean
    if normalize_to_0: 
        return intensity / intensity[0]
    else: return intensity
        
    
```
